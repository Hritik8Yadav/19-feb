```python

from flask import Flask

serv = Flask(__name__)


@serv.route("/")
def Hello():
    return"<h1> Hello World! <h1>"

if __name__=='__main__':
    serv.run(host="0.0.0.0")
```

```python
#Ans3: App routing is the process of mapping the url of a function so that it can be executed with the help of the url.
# That url will help the client to find the location of that code and execute it.
# App routing is generally used to execute the function remotely from any platform.
```

```python
from flask import Flask

serv = Flask(__name__)


@serv.route("/welcome")
def welcome():
    return"<h1> Welcome to ABC Corporation <h1>"

@serv.route("/")
def details():
    return '''Company Name: ABC Corporation \n
    Location: India \n
    Contact Detail: 999-999-9999'''

if __name__=='__main__':
    serv.run(host="0.0.0.0")
```

```python
#Ans5:- route decorator is used to bind the url. And url_for can also be used for dynamic binding of the url.
# for example:-
from flask import *

serv = Flask(__name__)


@serv.route("/welcome")
def welcome():
    return"<h1> Welcome to ABC Corporation <h1>"

@serv.route("/details")
def details():
    return '''Company Name: ABC Corporation \n
    Location: India \n
    Contact Detail: 999-999-9999'''
@serv.route("/<type>")
def dynamic_binding():
    if type=='welcome':
        return redirect(url_for(welcome))
    else:
        return redirect(url_for(details))

if __name__=='__main__':
    serv.run(host="0.0.0.0")
```
